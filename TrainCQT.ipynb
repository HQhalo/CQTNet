```python
from pydub import AudioSegment
import math

class SplitWavAudioMubin():
    def __init__(self, folder,outfolder, filename):
        self.folder = folder
        self.filename = filename
        self.filepath = folder + '/' + filename
        self.outfolder = outfolder
        self.audio = AudioSegment.from_wav(self.filepath)
    def get_duration(self):
        return self.audio.duration_seconds
    
    def single_split(self, from_min, to_min, split_filename, out):
        t1 = from_min  * 1000
        t2 = to_min  * 1000
        split_audio = self.audio[t1:t2]
        split_audio.export(out + '/' + split_filename, format="wav")
        
    def multiple_split(self, min_per_split):
        total_mins = int(self.get_duration())
        for i in range(0, total_mins, min_per_split):
            split_fn = self.filename.split('.')[0] + "_"+str(i) +'.wav'
            out = self.outfolder+"_" + str(i)
            if not os.path.exists(out):
              os.makedirs(out)
            self.single_split(i, i+min_per_split, split_fn, out)
            print(str(i) + ' Done')
            if i == total_mins - min_per_split:
                print('All splited successfully')
```

```python
from nnAudio import features
from scipy.io import wavfile
import torch
import numpy as np
SAMPLE_RATE = 16000
cqt_extractor = features.CQT(sr=SAMPLE_RATE).to('cuda')

def extractCqt(filename):
  sr, song = wavfile.read(filename) # Loading your audio
  assert sr == SAMPLE_RATE
  x = torch.tensor(song, device='cuda').float() # casting the array into a PyTorch Tensor
  spec = cqt_extractor(x)
  spec = spec[0].cpu().detach().numpy() 

  mean_size = 1

  height, length = spec.shape
  new_cqt = np.zeros((height,int(length/mean_size)),dtype=np.float64)
  for i in range(int(length/mean_size)):
      new_cqt[:,i] = spec[:,i*mean_size:(i+1)*mean_size].mean(axis=1)
  return new_cqt
```

```python
from tqdm import tqdm


for tk in tqdm(list(range(len(train_keys)))):
  for filename, version in zip(listFile[train_keys[tk]], tracks[train_keys[tk]]):
    new_cqt =  extractCqt(filename)
    np.save(f'train_npy/{tk}-{version}.npy', new_cqt)
    
for tk in list(range(len(val_keys))):
  for filename, version in zip(listFile[val_keys[tk]], tracks[val_keys[tk]]):
    if (version.split('_')[0] == 'hum' and version.split('_')[1] != 'gen') or version.split('_')[1] == 'vocals':
      new_cqt =  extractCqt(filename)
      np.save(f'val_npy/{tk}-{version}.npy', new_cqt)
    else: 
      print('skip ' + filename)
```

```python
from scipy.io import wavfile
import glob
import soundfile
import numpy
import random
import os

def loadWAV(filename, max_audio, in_sample_rate):
    audio, sample_rate = soundfile.read(filename)
    assert in_sample_rate == sample_rate
    audiosize = audio.shape[0]

    if audiosize <= max_audio:
        shortage    = max_audio - audiosize + 1 
        audio       = numpy.pad(audio, (0, shortage), 'wrap')
        audiosize   = audio.shape[0]


    startframe = numpy.array([numpy.int64(random.random()*(audiosize-max_audio))])
    
    feats = []

    for asf in startframe:
        feats.append(audio[int(asf):int(asf)+max_audio])

    feat = numpy.stack(feats,axis=0).astype(numpy.float)

    return feat;
    
class AugmentWAV(object):

    def __init__(self, musan_path):

        self.noisetypes = ['noise','speech','music']

        self.noisesnr   = {'noise':[0,15],'speech':[13,20],'music':[5,15]}
        self.numnoise   = {'noise':[1,1], 'speech':[3,7],  'music':[1,1] }
        self.noiselist  = {}

        augment_files   = glob.glob(os.path.join(musan_path,'*/*/*.wav'));
        print(augment_files)
        for file in augment_files:
            if not file.split('/')[-3] in self.noiselist:
                self.noiselist[file.split('/')[-3]] = []
            self.noiselist[file.split('/')[-3]].append(file)


    def additive_noise(self, noisecat, audio, sample_rate):

        clean_db = 10 * numpy.log10(numpy.mean(audio ** 2)+1e-4) 

        numnoise    = self.numnoise[noisecat]
        noiselist   = random.sample(self.noiselist[noisecat], random.randint(numnoise[0],numnoise[1]))

        noises = []

        for noise in noiselist:

            noiseaudio  = loadWAV(noise, audio.shape[0], sample_rate)
            noise_snr   = random.uniform(self.noisesnr[noisecat][0],self.noisesnr[noisecat][1])
            noise_db = 10 * numpy.log10(numpy.mean(noiseaudio[0] ** 2)+1e-4) 
            noises.append(numpy.sqrt(10 ** ((clean_db - noise_db - noise_snr) / 10)) * noiseaudio)

        return numpy.sum(numpy.concatenate(noises,axis=0),axis=0,keepdims=True) + audio

    
```

```python
import glob
import os
from sklearn.model_selection import train_test_split

aug_filenames = []
for id in list(os.listdir('/content/dataset_wav')):
  aug_filenames.append(list(glob.glob(f'/content/dataset_wav/{id}/hum_????.wav'))[0])
  
_, aug_filenames = train_test_split(aug_filenames, test_size=0.6, random_state=42)
  
print(len(aug_filenames))
```

```python
import soundfile
augment = AugmentWAV('/content/musan')
for filename in aug_filenames:
  
  audio, sample_rate = soundfile.read(filename)
  aug_audio = augment.additive_noise('noise',audio, sample_rate)
 
  dir = '/'.join(filename.split('/')[:-1]).replace('dataset_wav', 'aug/dataset_wav')
  name = filename.split('/')[-1].replace('hum', 'hum_gen_aug')
  out = dir + '/' + name
  if not os.path.exists(dir):
    os.makedirs(dir)
  soundfile.write(out, aug_audio.flatten(), sample_rate)
  
```
